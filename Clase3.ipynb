```
plt.figure( figsize = (6,5), dpi = 100)
plt.plot(x, x**2, 'r*--', linewidth=2, label = 'x^2')
plt.plot(x, x**3, 'g.-', linewidth=2, label = 'x^3')
plt.xlabel('tiempo')
plt.ylabel('precio')
plt.legend()
plt.show()
```

### Variando K

```
# Como funciona enumerate
lista = [1,7,8, 9]
for i, m in enumerate(lista):
  print(i, m)
```

```
vecinos = np.arange(1, 50)
```

```
train_acc = np.empty(len(vecinos))
test_acc = np.empty(len(vecinos))
```

```
for i, k in enumerate(vecinos):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(X_train, y_train)

  train_acc[i] = knn.score(X_train, y_train)
  test_acc[i] = knn.score(X_test, y_test)
```

```
train_acc[:10]
```

```
test_acc[:10]
```

```
plt.figure(figsize= (7,6), dpi=100)
plt.plot(vecinos, test_acc, label = 'Test Acc')
plt.plot(vecinos, train_acc, label = 'Train Acc')
plt.legend()
plt.xlabel('K vecinos')
plt.ylabel('Accuracy')
plt.show()
```
